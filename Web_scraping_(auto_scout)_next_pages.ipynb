```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.autoscout24.com/lst/mercedes-benz/slr?sort=standard&desc=0&offer=J%2CU%2CO%2CD&ustate=N%2CU&cy=D&fregfrom=2005&atype=C"  
```

```python
articles_dict={}
article_no=0
i=1
while True:
    response=requests.get(url)
    data=response.content
    soup=BeautifulSoup(data,'html.parser')
    
    articles=soup.find_all('div', class_='cldt-summary-full-item-main')
    
    for article in articles:
        model=article.find('h2', class_='cldt-summary-makemodel sc-font-bold sc-ellipsis').text
        title_tag=article.find('h2', class_='cldt-summary-version sc-ellipsis')
        title=title_tag.text if title_tag else 'N/A'
        price_tag=article.find('span', class_='cldt-price sc-font-xl sc-font-bold')
        price=price_tag.text if price_tag else 'N/A'
        price_striped=price.strip()
        mileage_tag=article.find('li', {'data-type':'mileage'})
        mileage=mileage_tag.text if mileage_tag else 'N/A'
        mileage_striped=mileage.strip()
        power_tag=article.find_all('li')
        power=power_tag[2].text if power_tag else 'N/A'
        power_striped=power.strip()
        first_registration_tag=article.find('li', {'data-type':'first-registration'})
        first_registration=first_registration_tag.text if first_registration_tag else 'N/A'
        first_reg_striped=first_registration.strip()
        previous_owners_tag=article.find('li', {'data-type':'previous-owners'})
        previous_owners=previous_owners_tag.text if previous_owners_tag else 'N/A'
        prev_owners_striped=previous_owners.strip()
        transmission_type_tag=article.find('li', {'data-type':'transmission-type'})
        transmission_type=transmission_type_tag.text if transmission_type_tag else 'N/A'
        transmission_striped=transmission_type.strip()
        fuel_tag=article.find('li', class_='summary_item_no_bottom_line')
        fuel=fuel_tag.text if fuel_tag else 'N/A'
        fuel_striped=fuel.strip()
        link=article.find('a', {'data-item-name':'detail-page-link'}).get('href')
        link_striped='https://www.autoscout.com'+link
        
        article_no+=1
        articles_dict[article_no]=[model, title, price_striped, mileage_striped, power_striped, first_reg_striped, prev_owners_striped, transmission_striped, fuel_striped, link_striped]
        
        print('Model:', model, '\nTitle:', title, '\nPrice:', price.strip(), '\nMileage:', mileage.strip(), '\nPower:', power.strip(), '\nFirst registration:', first_registration.strip(), '\nPrevious owners:', previous_owners.strip(), '\nTransmission type:', transmission_type.strip(), '\nFuel type:', fuel.strip(), '\nLink: https://www.autoscout.com'+link, '\n-----')
    
    i+=1
    url_tag='https://www.autoscout24.com/lst/mercedes-benz/slr?&sort=standard&desc=0&offer=J%2CU%2CO%2CD&ustate=N%2CU&size=20&cy=D&fregfrom=2005&atype=C&page={}'.format(i)
    if i<=2:
        url=url_tag
    else:
        break
            
articles_df=pd.DataFrame.from_dict(articles_dict, orient='index', columns=['Model', 'Title', 'Price', 'Mileage', 'Power', 'First Registration', 'Previous owners', 'Transmission', 'Fuel', 'Link'])
```
